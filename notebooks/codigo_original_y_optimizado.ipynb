‚úÖ **PARTE 1 ‚Äî C√≥digo Original (Sin Optimizaci√≥n)** 

```python
import time

def es_primo(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

def buscar_primos(limite):
    primos = []
    for num in range(1, limite + 1):
        if es_primo(num):
            primos.append(num)
    return primos

# Medici√≥n de tiempo
inicio = time.time()
resultado_original = buscar_primos(100000)
fin = time.time()

tiempo_original = fin - inicio

tiempo_original
```

‚òëÔ∏è **PARTE 2: Codigo optimizado** ‚¨áÔ∏è

```python
import time
import numpy as np
import math

def es_primo_opt(n):
    if n < 2:
        return False
    limite = int(math.sqrt(n)) + 1
    for i in range(2, limite):
        if n % i == 0:
            return False
    return True

def buscar_primos_opt(limite):
    numeros = np.arange(1, limite + 1)
    primos = [n for n in numeros if es_primo_opt(n)]
    return primos

# Medici√≥n de tiempo
inicio = time.time()
resultado_opt = buscar_primos_opt(100000)
fin = time.time()

tiempo_opt = fin - inicio

tiempo_opt
```

- **Comparacion:**

```python
print("Tiempo original:", tiempo_original)
print("Tiempo optimizado:", tiempo_opt)
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.bar(['Original', 'Optimizado'], [tiempo_original, tiempo_opt])
plt.title('Comparaci√≥n de Tiempos de Ejecuci√≥n')
plt.ylabel('Segundos')
plt.show()
```

- **Para su distribucion**

```python
original_samples = np.random.normal(tiempo_original, tiempo_original*0.1, 100)
optimized_samples = np.random.normal(tiempo_opt, tiempo_opt*0.1, 100)

plt.hist(original_samples, alpha=0.5, label='Original')
plt.hist(optimized_samples, alpha=0.5, label='Optimizado')
plt.legend()
plt.show()
```
